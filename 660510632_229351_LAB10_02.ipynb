```python
# import module ที่ต้องใช้
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from datetime import datetime
```

```python
# ฟังก์ชันที่เปลี่ยน string ให้เป็น datetime
def parser(dates):
  return datetime.strptime(dates, '%Y-%m-%d')

# parse_dates ชื่อของคอลัมน์ที่จะเปลี่ยนให้เป็น datetime
# index_col ชื่อของคอลัมน์ที่จะให้เป็น index
# date_parser ฟังก์ชันที่เปลี่ยน string ให้เป็น datetime
data = pd.read_csv('CM_temp.csv', parse_dates=['Date'],
                                        index_col='Date',
                                        date_parser=parser)

data.tail()
```

ข้อมูลที่ได้เป็นข้อมูลรายวัน ทำการหาค่าเฉลี่ยรายเดือนด้วย method `resample`

```python
data_monthly = data['MaxTemp'].resample("M").mean()

data_monthly.tail()
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))
```

```python
print(data_monthly.index.inferred_freq)
```

```python
# แบ่ง train test
train = data_monthly.iloc[:-24]
test = data_monthly.iloc[-24:]

print(test)
```

```python
plt.figure(figsize=(12,6))
data_monthly.plot();
```

```python
#Holt Winter’s Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

es = ExponentialSmoothing(train, trend ='mul', damped=True, seasonal='add', seasonal_periods=12) #สร้าง Model
es_fit = es.fit(optimized=True) #run เร็วขึ้น
yhat = es_fit.predict('2017-08-31', '2019-07-31')
```

```python
print('RMSE:',rmse(test,yhat))
```

```python
# 1.trend=add, damped=True, seasonal=None, seasonal_periods=12 --> RMSE: 2.7206000811236914
# 2.trend = 'mul', damped=True, seasonal=None, seasonal_periods=12 -->RMSE: 2.7206000811236675
# 3.trend = 'add', damped=False, seasonal=None, seasonal_periods=12 -->RMSE: 2.697261490836458
# 4.trend = 'additive', damped=False, seasonal='mul', seasonal_periods=12 -->RMSE: 1.1639766564724707
# 5.trend = 'mul', damped=False, seasonal='add', seasonal_periods=12 -->RMSE: 1.1875960299599877
# 6.trend = 'add', damped=True, seasonal='add', seasonal_periods=12 -->RMSE: 1.1134955121933203
# 7.trend = 'mul', damped=True, seasonal='add', seasonal_periods=12 -->RMSE: 0.8958522762071295
# 8.trend = 'mul', damped=True, seasonal='mul', seasonal_periods=12 -->RMSE: 1.10643357265325
```

```python
print('RMSE:',rmse(test,yhat))

plt.figure(figsize=(12,6))
data_monthly.plot()
yhat.plot();
```
